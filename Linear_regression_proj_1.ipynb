In  this project we get a file containg info for HR manager on employees. Our task is to build a linear regression ML model and to train it on our dataset to predict a salary that fits a particular employee.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
!pip install word2number
from word2number import w2n
```

```python
df = pd.read_csv("hiring.csv")
df
```

## First, perform some data cleaning üßπ
We'll fill missing values using functions fillna and median.
Then, we'll transform numerical values that are stored as not numbers into number objects.

```python
df.experience = df.experience.fillna('zero')
df
```

```python
median_test_score = math.floor(df['test_score(out of 10)'].median())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
df['experience'] = df['experience'].astype("string")
df['experience'].dtype
df
```

```python
df.experience=df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
df.columns
```

## Now we have a dataset that is ready to be fed into ML model ü¶æ
We'll build our linear regression model and fit our data into it.

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Prediction trial üîÆ
We'll test our model by predicting salaries for 2 new employees.
First employee has 2 yrs exp, 9 test score and 6 interview score.
Second employee has 12 yrs exp, 10 test score and 10 interview score.

```python
reg.predict([[2, 9, 6]])
```
